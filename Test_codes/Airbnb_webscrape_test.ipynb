```python
def get_room_sites(websites):
    
    """This function takes a list of websites from AirBNB 
    uses Selenium to open each website and parse out the 
    innerHTML files. The innerHTML file is then used by 
    Beautiful Soup to find all the a tags and recursively 
    go over them to remove the href tags within it. The 
    function returns a list with all duplicates removed."""
    
    chromedriver = '/Applications/chromedriver'
    os.environ['webdriver.chrome.driver'] = chromedriver
    driver = webdriver.Chrome(chromedriver)
    time.sleep(1)
    
    room_list = []

    for website in websites:
        driver.get(website)
        time.sleep(1)
        innerHTML = driver.execute_script('return document.body.innerHTML')
        soup = BeautifulSoup(innerHTML, 'lxml')
        s=soup.find_all('a')
        for i,v in enumerate(s):
            m = s[i]['href']
            room_list.append(m)


    driver.quit()
    return list(set(room_list))
```

```python
def extract_fields(html_file):
    innerHTML1=open(html_file)
    soup = BeautifulSoup(innerHTML1, 'lxml')
    s = soup.find_all('span')
    dataframe = pd.DataFrame(columns = ['Desc', 'Total_Reviews', 'Guests', 'Rental_Type', 'Number_of_Beds', 'Number_of_Baths','Cost_per_night'], index =[0])
    dict1={}
    dict1['Desc'] = [s[15].text]
    dict1['Total_Reviews'] = [s[14].text]
    dict1['Guests'] = [s[22].text.split(' ')[0]]
    dict1['Rental_Type'] = [s[23].text]
    dict1['Number_of_Beds'] = [s[24].text.split(" ")[0]]
    dict1['Number_of_Baths'] = [s[25].text.split(" ")[0]]
    dict1['Cost_per_night'] = [s[74].text]
    dataframe = pd.DataFrame.from_dict(dict1)
    
    return dataframe
```
